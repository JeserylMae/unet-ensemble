## 00-1 Fine Tuning Stable Diffusion With LoRA

Mount Drive & Import Libraries

```python
import os
from PIL import Image
import torch
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader
from transformers import CLIPTokenizer, CLIPTextModel
from diffusers import AutoencoderKL, UNet2DConditionModel, DDPMScheduler
from peft import LoraConfig, get_peft_model
from accelerate import Accelerator
```

### Dataset Preparation

```python
save_dir = "/content/drive/Shareddrives/U-Net Ensemble/Dataset/ffhq_5000"
dataset_with_prompts = []

image_files = sorted(os.listdir(save_dir))
for filename in image_files:
    if filename.lower().endswith((".png", ".jpg", ".jpeg")):
        filepath = os.path.join(save_dir, filename)
        dataset_with_prompts.append({
            "imagepath": filepath,
            "text": "a photo of a person"
        })
```

Split dataset (85% train, 15% val)

```python
split_idx = int(0.85 * len(dataset_with_prompts))
train_data = dataset_with_prompts[:split_idx]
val_data   = dataset_with_prompts[split_idx:]
```

### Dataset Preprocessing

Tokenizer & Image Transform

```python
model_id = "stable-diffusion-v1-5/stable-diffusion-v1-5"
tokenizer = CLIPTokenizer.from_pretrained(model_id, subfolder="tokenizer")

image_transform = transforms.Compose([
    transforms.Resize((512, 512)),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])
```

Lazy Dataset Class

```python
class LazyFFHQDataset(Dataset):
    def __init__(self, data, tokenizer, transform):
        self.data = data
        self.tokenizer = tokenizer
        self.transform = transform

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        item = self.data[idx]
        image = Image.open(item["imagepath"]).convert("RGB")
        pixel_values = self.transform(image)
        input_ids = self.tokenizer(
            item["text"],
            padding="max_length",
            truncation=True,
            max_length=self.tokenizer.model_max_length,
            return_tensors="pt"
        ).input_ids[0]
        return {"pixel_values": pixel_values, "input_ids": input_ids}
```

```python
train_dataset_lazy = LazyFFHQDataset(train_data, tokenizer, image_transform)
val_dataset_lazy   = LazyFFHQDataset(val_data, tokenizer, image_transform)
```

```python
train_dataloader = DataLoader(
    train_dataset_lazy, batch_size=2, shuffle=True, num_workers=2
)
val_dataloader = DataLoader(
    val_dataset_lazy, batch_size=2, shuffle=False, num_workers=2
)
```

### Pre-training set up

Load Stable Diffusion Components

```python
# device = "cuda" if torch.cuda.is_available() else "cpu"
accelerator = Accelerator(mixed_precision="fp16")
device = accelerator.device

vae = AutoencoderKL.from_pretrained(model_id, subfolder="vae").to(device)
unet = UNet2DConditionModel.from_pretrained(model_id, subfolder="unet").to(device)
text_encoder = CLIPTextModel.from_pretrained(model_id, subfolder="text_encoder").to(device)

vae.requires_grad_(False)
text_encoder.requires_grad_(False)
unet.requires_grad_(False)  # LoRA will add trainable layers
```

```python
optimizer = torch.optim.AdamW(unet.parameters(), lr=1e-4)

unet, optimizer, train_dataloader, val_dataloader = accelerator.prepare(
    unet, optimizer, train_dataloader, val_dataloader
)

vae.to(device)
text_encoder.to(device)
```

LoRA Setup

```python
lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    target_modules=["to_q", "to_k", "to_v", "to_out.0"],
    lora_dropout=0.05,
    bias="none",
)
unet = get_peft_model(unet, lora_config)
unet.train()

optimizer = torch.optim.AdamW(unet.parameters(), lr=1e-4)
trainable_params = sum(p.numel() for p in unet.parameters() if p.requires_grad)
print("Trainable params (LoRA):", trainable_params)
```

Noise Scheduler

```python
noise_scheduler = DDPMScheduler.from_pretrained(model_id, subfolder="scheduler")
```

### Training Loop with Mixed Precision

```python
from torch.amp import autocast, GradScaler

scaler = GradScaler("cuda")

num_epochs = 1000
best_val_loss = float('inf')
patience = 5
counter = 0
modelctr = 0

for epoch in range(num_epochs):
    unet.train()
    train_loss = 0.0

    for batch in train_dataloader:
        pixel_values = batch["pixel_values"].to(device)
        input_ids = batch["input_ids"].to(device)

        optimizer.zero_grad(set_to_none=True)

        with torch.no_grad():
            latents = vae.encode(pixel_values).latent_dist.sample()
            latents = latents * 0.18215
            noise = torch.randn_like(latents)
            timesteps = torch.randint(
                0,
                noise_scheduler.config.num_train_timesteps,
                (latents.shape[0],),
                device=latents.device
            )
            noisy_latents = noise_scheduler.add_noise(latents, noise, timesteps)
            encoder_hidden_states = text_encoder(input_ids)[0]

        with accelerator.autocast():
            noise_pred = unet(
                noisy_latents, timesteps, encoder_hidden_states
            ).sample
            loss = torch.nn.functional.mse_loss(noise_pred, noise)

        accelerator.backward(loss)
        optimizer.step()

        train_loss += loss.item()

    train_loss /= len(train_dataloader)

    # --------- VALIDATION ---------
    unet.eval()
    val_loss = 0.0
    with torch.no_grad():
        for batch in val_dataloader:
            pixel_values = batch["pixel_values"].to(device)
            input_ids = batch["input_ids"].to(device)

            latents = vae.encode(pixel_values).latent_dist.sample()
            latents = latents * 0.18215
            noise = torch.randn_like(latents)
            timesteps = torch.randint(
                0,
                noise_scheduler.config.num_train_timesteps,
                (latents.shape[0],),
                device=device
            )
            noisy_latents = noise_scheduler.add_noise(latents, noise, timesteps)
            encoder_hidden_states = text_encoder(input_ids)[0]

            with accelerator.autocast():
                noise_pred = unet(
                    noisy_latents, timesteps, encoder_hidden_states
                ).sample
                loss = torch.nn.functional.mse_loss(noise_pred, noise)

            val_loss += loss.item()

    val_loss /= len(val_dataloader)

    print(
        f"Epoch [{epoch+1}/{num_epochs}] "
        f"Train Loss: {train_loss:.6f} | "
        f"Val Loss: {val_loss:.6f}"
    )

    if val_loss < best_val_loss:
        best_val_loss = val_loss
        counter = 0

        accelerator.wait_for_everyone()
        if accelerator.is_main_process:
            unwrapped_unet = accelerator.unwrap_model(unet)
            torch.save(
                unwrapped_unet.state_dict(),
                f"/content/drive/MyDrive/checkpoints/best_model-{modelctr}.pt"
            )
            modelctr += 1
    else:
        counter += 1
        if counter >= patience:
            print("Early stopping triggered.")
            break
```

```python
from pydantic import BaseModel
import torch
from io import BytesIO
import base64
from typing import List, Optional
from io import BytesIO
from diffusers import AutoencoderKL, UNet2DConditionModel, StableDiffusionPipeline, DDPMScheduler
from transformers import CLIPTokenizer, CLIPTextModel
from peft import LoraConfig, get_peft_model
import torch
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model_id = "stable-diffusion-v1-5/stable-diffusion-v1-5"

# Load base models
vae = AutoencoderKL.from_pretrained(model_id, subfolder="vae").to(device)
text_encoder = CLIPTextModel.from_pretrained(model_id, subfolder="text_encoder").to(device)
tokenizer = CLIPTokenizer.from_pretrained(model_id, subfolder="tokenizer")

# Load UNet + LoRA
unet = UNet2DConditionModel.from_pretrained(model_id, subfolder="unet").to(device)
lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    target_modules=["to_q", "to_k", "to_v", "to_out.0"],
    lora_dropout=0.05,
    bias="none",
)

unet = get_peft_model(unet, lora_config)
unet.load_state_dict(torch.load(model_path), strict=False)
unet.eval()

# Scheduler
scheduler = DDPMScheduler.from_pretrained(model_id, subfolder="scheduler")
```

```python
pipe = StableDiffusionPipeline(
    vae=vae,
    text_encoder=text_encoder,
    tokenizer=tokenizer,
    unet=unet,
    scheduler=scheduler,
    safety_checker=None,
    feature_extractor=None
).to(device)

pipe.enable_attention_slicing()
```
